```python
import pandas as pd


# careplans = pd.read_csv("careplans.csv")
claims = pd.read_csv("claims.csv")
claims_transactions = pd.read_csv("claims_transactions.csv")
# conditions = pd.read_csv("conditions.csv")
# devices = pd.read_csv("devices.csv")

# imaging_studies = pd.read_csv("imaging_studies.csv")
observations = pd.read_csv("observations.csv")
# medications = pd.read_csv("medications.csv")
patients = pd.read_csv("patients.csv")
```

```python
claims = claims.rename({"Id": "Claims_ID"}, axis=1)
```

```python
merged_df = patients.merge(claims, left_on="Id", right_on="PATIENTID", how="right")
```

```python
claims_transactions_grouped = (
    claims_transactions.groupby(["CLAIMID", "PATIENTID"])["NOTES"]
    .apply(lambda x: ", ".join(map(str, x)))
    .reset_index()
)
claims_transactions_grouped
```

```python
merged_df.columns
```

```python
merged_df = merged_df.merge(
    claims_transactions_grouped,
    left_on=["Id", "Claims_ID"],
    right_on=["PATIENTID", "CLAIMID"],
    how="inner",
)
```

```python
merged_df["STATUS1"] = np.where(merged_df["STATUS1"] == "CLOSED", 0, 1)
merged_df["STATUS2"] = np.where(merged_df["STATUS2"] == "CLOSED", 0, 1)
merged_df["STATUSP"] = np.where(merged_df["STATUSP"] == "CLOSED", 0, 1)
```

```python
merged_df["Claim_Status_Decision"] = merged_df[["STATUS1", "STATUS2", "STATUSP"]].max(
    axis=1
)
```

```python
merged_df = merged_df.drop(["STATUS1", "STATUS2", "STATUSP"], axis=1)
```
